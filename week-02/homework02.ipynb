````python
# sample solution
def polynomial(x, coefficients):
    '''calculate the value of a polynomial function of degree 3
    from a list of exxactly 4 coefficients'''

    return x ** 3 * coefficients[0] + x ** 2 * coefficients[1] + x * coefficients[2] + coefficients[3]


````

**Problem 10**

**Concept:** *Function that returns a list.*

**Task:**

Write a function named ```first_mid_last(my_list)``` that:

* Takes a list named ```my_list```

* Returns a new list containing:

    * First element

    * Middle element

    * Last element

**Note:**

For lists with even number of elements, you should return the second middle element. See sample test cases for details.

**Sample Output:**

```
first_mid_last([1, 2, 3, 4, 5]) = [1, 3, 5]
first_mid_last([6,1,7,3,5,8,2,3,6,1]) = [6, 8, 1]
```

````python
# sample solution
def first_mid_last(my_list):
    '''returns the first, middle, and last elenets of `my_list`.'''

    first = my_list[0]
    middle = my_list[len(my_list) // 2]
    last = my_list[-1]

    return [first, middle, last]
````

**Problem 11**

**Concept:** *Function that returns a list.*

**Task:**

Write a function named ```change_from_a_dollar(price)``` that:

* Takes an integer ```price``` (in cents, less than 100)

* Calculates how much change is needed from one dollar (100 cents)

* Returns a list in the following order:

    1. Number of quarters (1 quarter = 25 cents)

    2. Number of dimes (1 dime = 10 cents)

    3. Number of nickels (1 nickel = 5 cents)

    4. Number of pennies (1 penny = 1 cent)

**Sample Output:**

```
change_from_a_dollar(22) returned [3, 0, 0, 3]
change_from_a_dollar(42) returned [2, 0, 1, 3]
```

````python
# sample solution
def change_from_a_dollar(price):
    '''return the number of each coin required to make change from a dollar.
    we assume a price in cents (integer) less than 100.
    returns a list of integers corresponding to the number of each coin
    (quarters, dimes, nikels, pennies).'''

    # compute total amount of change
    change = 100 - price
    # compute quantity of each coin, and how much change is left after that coin
    quarters = change // 25
    change = change % 25
    dimes = change // 10
    change = change % 10
    nickels = change // 5
    pennies = change % 5

    return [quarters, dimes, nickels, pennies]
````

**Test cases for problem 11**

````python
print("change_from_a_dollar(22) returned", change_from_a_dollar(22))
print("change_from_a_dollar(42) returned", change_from_a_dollar(42))
````
